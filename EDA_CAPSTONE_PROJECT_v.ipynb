```python
## first checking for duplicvate rows if any
hb.duplicated().value_counts()
## total number of duplicate row are : 5932
plt.figure(figsize=(5,5))
sns.countplot(x=hb.duplicated())
```

# **Handling missing values in the given data**

*   company and agent has their own data,sometimes customer didnt booked hotel through any company or agent. so,in that case values can be replaced by 0.



```python
#checking for null values using .isnull()
hb.isnull().sum().sort_values(ascending = False).head(6)
```

```python
#we will replace null values in company &agent to 0
hb[['company','agent']] = hb[['company','agent']].fillna(0)
```

```python
# in colom childfen we found value is 0 ,which means there no children were present in the group of coustomer
## we replaced the null values in children coloumn with mean().
hb['children'].fillna(hb['children'].mean(), inplace= True)
```

```python
# missing values in country coloum is replaced by others.
hb['country'].fillna('others', inplace= True)
```

```python
hb.isnull().sum().sort_values(ascending = False).head(8)
```

```python
# there are some rows of children,babies& adults with value 0
hb[hb['children']+hb['babies']+hb['adults'] == 0].shape
```

```python
#as we noticed values 0 for children,babies &adults we removed 0 values using .drop
hb.drop(hb[hb['children']+hb['babies']+hb['adults'] ==0].index, inplace= True)
```

#**Converting DataTypes of columns**

```python
## we convetrting data types of children, company & agent from float to int,because those cant be in float values
hb[['children','company','agent']] = hb[['children','company','agent']].astype('int64')
```

```python
hb.T
```

```python
hb['reservaion_status_date'] = pd.to_datetime(hb['reservation_status_date'], format = '%Y-%m-%d')
```

```python
hb.T
```

```python
#creating the new coulms so that we can analyse the stay length at respective hotel
## adding total_stays of customers in hotel
hb['total_stay'] = hb['stays_in_week_nights'] + hb['stays_in_weekend_nights']

## adding total_people colums from combining children,babies & adults 
hb['total_people'] = hb['children'] + hb['babies'] + hb['adults']
```

```python
# now we can see that total_people & total_stay has been added
hb.T
```
